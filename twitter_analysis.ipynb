```python
%matplotlib qt
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
```

```python
df = pd.read_csv("cleaned_twitter_data.csv", encoding = "ISO-8859-1", parse_dates = ["date"], index_col="date")
df.head()
```

```python
df.describe()
```

```python
len(df)
```

```python
mn_df = df.groupby(pd.Grouper(freq="M")).mean()
mn_df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
(mn_df["polarity"] - mn_df["polarity"].mean()).plot(ax=ax, xticks=mn_df.index)
ax.set_xlabel("Date")
ax.set_ylabel("Polarity")
ax.set_title("Polarity Decreases (and Remains Decreased) After Snowden")
plt.axvline(x=dt.datetime(2013, 6, 6), lw=3.0)
plt.axhline(y=0, lw=3.0, ls='--')
#plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
(mn_df["subjectivity"] - mn_df["subjectivity"].mean()).plot(ax=ax, xticks=mn_df.index)
ax.set_xlabel("Date")
ax.set_ylabel("Subjectivity")
plt.axvline(x=dt.datetime(2013, 6, 6), lw=3.0)
plt.axhline(y=0, lw=3.0, ls='--')
#plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
(mn_df["retweets"]).plot(ax=ax, xticks=mn_df.index)
ax.set_xlabel("Date")
ax.set_ylabel("Retweets")
plt.axvline(x=dt.datetime(2013, 6, 6), lw=3.0)
plt.axhline(y=0, lw=3.0, ls='--')
#plt.grid()
plt.show()
```
